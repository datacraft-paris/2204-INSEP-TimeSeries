### Renommer les colonnes et encoder df_morning

```python
def rename_df_morning(df):
    df.rename(columns={"Date":"date","About":"about","A quelle heure je me suis couchee :":"heure_couche",
                       "A quelle heure je me suis levee :":'heure_leve',"Heures de sommeil":"nb_heure_sommeil",
                       "Qualité du sommeil":"qualite_sommeil",
                      "Comment te sens tu ce matin ?":"ressentit_global",
                       "Ce matin je ressens des émotions/sentiments plutôt :":"ressentit_emotion",
                       "Je ressens, ce matin (ou la veille) un ou plusieurs de ces symptôme(s) :":"symptomes",
                       "Quel est ton poids aujourd'hui ?":"poids",
                       "Une (ou des) blessures/douleurs/gène importantes à signaler ?":"blessures_douleurs",
                       "Zones de douleur/gène":"zone_de_douleur",
                       "Douleur/gène":"douleur_gene",
                       "Intensité douleur/gène":"intensite_douleur",
                       "Type Douleur/Gène":"type_douleur"
                      }, inplace=True)
```

```python
def encode_morning_ordinal(df): # peut-être une version one-hot encodé serait judicieuse
    #symptomes
    le = preprocessing.LabelEncoder()
    le.fit(list(df['Je ressens, ce matin (ou la veille) un ou plusieurs de ces symptôme(s) :'].unique()))
    df['symptomes_enc']=le.transform(df['Je ressens, ce matin (ou la veille) un ou plusieurs de ces symptôme(s) :'])
    
    #blessure
    df['Une (ou des) blessures/douleurs/gène importantes à signaler ?']=df['Une (ou des) blessures/douleurs/gène importantes à signaler ?'].fillna('Aucune')    
    le = preprocessing.LabelEncoder()
    le.fit(list(df['Une (ou des) blessures/douleurs/gène importantes à signaler ?'].unique()))
    df['blessures_enc']=le.transform(df['Une (ou des) blessures/douleurs/gène importantes à signaler ?'])
    
    #zone de douleur
    df['Zones de douleur/gène']=df['Zones de douleur/gène'].fillna("Aucune")    
    le = preprocessing.LabelEncoder()
    le.fit(list(df['Zones de douleur/gène'].unique()))
    df['zone_douleur_enc']=le.transform(df['Zones de douleur/gène'])
    
    #douleur
    df['Douleur/gène']=df['Douleur/gène'].fillna("Aucune")    
    le = preprocessing.LabelEncoder()
    le.fit(list(df['Douleur/gène'].unique()))
    df['douleur_enc']=le.transform(df['Douleur/gène'])
    
    #type de douleur
    df_morning_ath1_vf['Type Douleur/Gène']=df['Type Douleur/Gène'].fillna("Aucun")    
    le = preprocessing.LabelEncoder()
    le.fit(list(df['Type Douleur/Gène'].unique()))
    df['type_douleur_enc']=le.transform(df['Type Douleur/Gène'])
```

```python
encode_morning_ordinal(df_morning_ath1_vf)
encode_morning_ordinal(df_morning_ath2_vf)
```

```python
rename_df_morning(df_morning_ath1_vf)
rename_df_morning(df_morning_ath2_vf)
```
